```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data = np.array([250,275,280,300,320,325,330,350,360,10000])
```

```python
z_scores = abs(stats.zscore(data))
```

```python
threshold =3
```

```python
outliers = np.where(z_scores > threshold)[0]
print("Outliers(indices):", outliers)
```

```python
colors = "red"
```

```python
plt.figure(figsize=(8,6))
plt.scatter(np.arange(len(data)), data, c=colors)
plt.xlabel("Data Point Index")
plt.ylabel("Sale price (in $000s)")
plt.title("Outliers Detection using zscore")
plt.show()
```
